```python
#Import appropriate code
import pandas as pd
```

```python
#Import uncleaned csv
file = 'House_price_multifeatures.csv'
```

```python
#Read csv and present as DataFrame
df = pd.read_csv(file)
```

```python
#Create small dataframe for experimentation. Decreased wait time for the large set of data
df_short = df.head(10)
```

```python
#Display Dataframe
df_short
```

```python
#Create filtered DataFrame with applicable columns

new_df = df[['Id','OwnerCity', 'OwnerState','PropertyLocationStreet1','PropertyCity','PropertyZip','StateParcelId','ParcelSize','ValueofHome','YearBuilt','TypeOfResidence','NumberOfStories','AdjSquareFeet',
            'Pool','PPC', 'BCEG', 'DistancetoCoast', 'PoliceID','Crime','FireID','DistancetoFireDepartment', 'DistancetoSinkhole','SinkholeLatitude',
            'SinkholeLongitude', 'LocationWindSpeed', 'WindborneDebrisRegion', 'Terrain', 'FloodZone', 'FloodZoneRiskLevel']]
new_df
```

```python
#Use unique function to identify potential duplicates. Ex: FL and Florida
new_df['OwnerState'].unique()
```

```python
#Replace Full State names with abbreviations

new_df['OwnerState'] = new_df['OwnerState'].replace(
    {'NORTH CAROLINA': 'NC', 'VIRGINIA': 'VA', 'SOUTH CAROLINA': 'SC'})
```

```python
#Check to see if code worked and "North Carolina" is no longer in data, but NC is.
new_df['OwnerState'].unique()
```

```python
#Filter DataFrame to only include NC, SC and VA. Display DataFrame

clean_df = new_df.loc[(new_df['OwnerState'] == 'NC') | (new_df['OwnerState'] == 'SC') | (new_df['OwnerState'] == 'VA')]
clean_df
```

```python
#Drop Duplicates
clean_df.drop_duplicates(subset=['Id'])
```

```python
#Export to CSV
clean_df.to_csv('clean.csv', index = False, header = True)
```
